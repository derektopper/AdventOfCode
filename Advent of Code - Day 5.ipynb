```python
max_val = 0
for i in lst:
    rang = list(range(128))
    for letter in i[0:7]:
        if letter == "F":
            rang = rang[:len(rang)//2]
        if letter == "B":
            rang = rang[len(rang)//2:]
    row = rang[0]
    
    rang = list(range(8))
    for letter in i[7:11]:
        if letter == "L":
            rang = rang[:len(rang)//2]
        if letter == "R":
            rang = rang[len(rang)//2:]
    col = rang[0]
    
    val = (row*8)+col
    if val > max_val:
        max_val = val
max_val

```

```python
aocd.submit(max_val, part='a', day=5, year=2020)
```
